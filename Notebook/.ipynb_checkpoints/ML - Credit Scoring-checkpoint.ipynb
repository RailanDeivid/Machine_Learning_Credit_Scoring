## Separando Variáveis

 A base de dados que estamos trabalhando até agora é inspirada em um conjunto de dados alemão e, como podemos perceber, temos uma única variável que reúne informações de sexo e de estado civil.  Acontece aqui, no Brasil, os modelos de créditos são fiscalizados e regulamentados pelo Banco Central e existem diversas regras que precisam ser seguidas e existem informações que não podem ser associadas ao modelo.
 
 
 Essas informações são as chamadas informações sensíveis. O que nós consideramos aqui, no Brasil, o que o Banco Central considera uma informação sensível? Origem social e étnica, condições de saúde, informação genética, orientação sexual, convicções políticas, religiosas e filosóficas.

Então o sexo é uma informação sensível e o estado civil não é. Pelo Banco Central, nós não podemos fazer a diferenciação na hora da concessão de crédito para diferentes sexos. Cada empresa pode decidir políticas de crédito diferentes, mas a concessão, o modelo de concessão, não pode ter como argumento o sexo. Voltando para a nossa base de dados, nós temos usado até agora o sexo e estado civil aqui.


**E diante disso, iremos tratar essas variáveis.**

* Categorias da variável 'sexo_est_civil'
        male: divorced/separated = 1
        female: divorced/separated/married = 2
        male: single = 3
        male: married/widowed = 4
        female: single =5
        
* Nova variável 'sexo'
        1 = mulher
        0 = homem
        
* Nova variável 'estado_civil'
        1 = solteira
        0 = casada/divorciada/viúva

E para esse tratamento usarei a função `map()` novamente.

* **Criando a variável 'sexo'**

```python
# setando um dic para as alterações
dic_sexo = {2:1,
            5:1,
            1:0,
            3:0,
            4:0}

# aplicando a função
df['sexo'] = df['sexo_est_civil'].map(dic_sexo)

# visualizando resultado
df[['sexo_est_civil', 'sexo']].head()
```

* **Criando a variável 'estado_civil'**

```python
# setando um dic para as alterações
dic_estado_civil = {3:1,
                    5:1,
                    1:0,
                    2:0,
                    4:0}

# aplicando a função
df['estado_civil'] = df['sexo_est_civil'].map(dic_estado_civil)

# visualizando resultado
df[['sexo_est_civil', 'estado_civil']].head()
```

## Descartando colunas

```python
colunas_removidas = ['sexo_est_civil', 'sexo', 'idade', 'prazo_emprestimo_meses', 'valor_emprestimo']
dados = df.drop(columns=colunas_removidas)
dados.head()
```

# Criando bases de treino e teste

```python
y = dados['default']
x = dados.drop(columns='default')
```

```python
from sklearn.model_selection import train_test_split

SEED = 77
treino_x, teste_x, treino_y, teste_y = train_test_split(
    x, y, test_size=0.30, random_state=SEED)
```

```python
print('A base de treino contém {} elementos e a base de teste contém {} elemento.'.format(
    treino_x.shape[0], teste_x.shape[0]))
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter=1000)
modelo.fit(treino_x, treino_y)

# acuracia de treino
print(modelo.score(treino_x, treino_y))
```

# Verificando as Métricas

 cross-validation é uma técnica para para saber quão bem o seu modelo está sendo generalizado pelo modelo supervisionado (a regressão logística é um desses modelos) produzido. Você pode se perguntar se a separação que fizemos da base em teste já não cumpre esta função. E a resposta é que sim, a base de teste já serve como avaliação para o seu modelo.

Entretanto, o cross-validation cria diferentes amostras da base de dados para treino e teste e assim, você pode ter certeza da sua performance. Existem diversas técnicas para fazer isso, mas aqui, vamos focar apenas em uma técnica: o K-Fold.

Esta técnica consiste em dividir sua base K vezes, realizar o treino com a maior parte destas amostras e depois testar a última parte, geralmente.

## Acurácia de teste

```python
previsoes = modelo.predict(teste_x)
previsoes
```

ele me devolve um array de 0 e 1. O que significa isso?

Significa que ele pegou a nossa base de teste e, a partir do modelo, ele me disse se aquela base terá default 0 ou default 1. Essas previsões serão usadas como parâmetro para calcularmos, finalmente, a nossa acurácia. Eu vou usar, para calcular a acurácia, o accuracy score, que é uma função pré-programada do Scikit-learn.

```python
from sklearn.metrics import accuracy_score

acuracia = round(accuracy_score(teste_y, previsoes), 2) *100
acuracia
```

 ## Plotando a matriz de confusão

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
```

```python
plt.style.use('default')
matriz_confusao = plot_confusion_matrix(modelo, teste_x, teste_y, cmap='Blues')
```

**Entendendo a matriz de confusão**

Matriz de Confusão.

De um modo mais simples a a Matriz de Confusão é calculada pela função de pesquisa Classificação. Ela exibe a distribuição dos registros em termos de suas classes atuais e de suas classes previstas. Isso indica a qualidade do modelo atual. Um modelo pode conter duas ou mais classes previstas.

é uma ferramenta muito usada para entender e avaliar mais a fundo a assertividade do seu modelo.

## Classification report

```python
from sklearn.metrics import classification_report

print(classification_report(teste_y, previsoes))
```

* **Precision:**

     Proporção de predições corretas de uma categoria em relação a todas as predições desta categoria.
         Precisão = VP / VP + FP
     
     Ou seja, de todos os dados avaliados quantos relamente o modelo acertou.
     
         
* **Recall:**
    
    Proporção de predições corretas de categoria alvo.
        Recall = VP / VP + FN
    
* **F1-Score:**

    Média harmônica entre precisão e recall.
        F1 - score = 2 * (Precisão * Recall) / (Precisão + Recall)


* **Support:**

    Quantidade absoluta de casos em cada categoria.
        

## Plotando a curva ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score

prob_previsao = modelo.predict_proba(teste_x)[:, 1]

prob_previsao
```

ele nos trás uma array com a probabilidade do meu evento ser predito como 1 ou como 0. Isso será importante e será necessário como argumento para calcular a nossa curva ROC.

* Setando os parametros para a plotagem da curva ROC.

```python
#tfp = taxa de falsos positivos
#tvp = taxa de verdadeiros positivos

tfp, tvp, limite = roc_curve(teste_y, prob_previsao)


""" 
    A roc curve vai desenhar, vai definir qual será a minha curva ROC.
    A roc auc score calcula qual é a área sob a curva ROC. 
    Ele calcula, através de uma integral, qual será a minha área sob a minha curva ROC.

"""
 
print('                   roc_auc', round(roc_auc_score(teste_y, prob_previsao),1))

import seaborn as sns
plt.style.use('seaborn')
# Configurações do layout
fig, ax = plt.subplots(figsize=(5, 5))

sns.lineplot(tfp, tvp)
# setando titulo do gráfico
ax.set_title("Cruva ROC.", size=14)
# setando titulo eixo x
ax.set_xlabel("Especifidade", fontsize=14)
# setando titulo eixo y
ax.set_ylabel("Sensibilidade", fontsize=14)

plt.plot([0, 1], ls="--", c = 'red') #plotando linha guia pontilhada vermelha
plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') #plotando linha guia pontilhada verde


fig.tight_layout()
```

* area > 0.8, entao discriminacao excelente!

A curva ROC Avalia o desempenho do modelo a partir da relação entre a taxa de verdadeiros positivos (sensibilidade) e a taxa de falsos positivos (especificidade) para diferentes limites.


Especificidade = VN / (VN + FP)  
Sensibilidade = VP (VP + FN)  

  O que significam essas duas métricas? 
          
   A sensibilidade indica a taxa de casos em que a categoria alvo, ou seja, a categoria 1, foi classificada corretamente.  
   Já a especificidade mede a taxa em que a categoria 0 foi, de maneira errada, classificada como 1


* **AUC:**

     Quantifica a Curva ROC através da área sob esta curva

    Convenção Estatística:

    - área < 0,7: baixa discriminiação
    - 0,7 ≤ área < 0,8: discriminação aceitável
    - 0,8 ≤ área < 0,9: discriminação excelente
    - área > 0,9: discriminação excepcional

# Ajuste do modelo

* Criando um função para modelagem

```python
def modelo(dados, classificador):  
# separando base de modelagem e variável resposta
    y = dados['default']
    x = dados.drop(columns='default')
# separando dados de treino e teste
    SEED = 77
    treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size=0.30, random_state=SEED)
    base_treino = treino_x.shape[0]
    base_teste =  teste_x.shape[0]
    print('A base de treino contém {} elementos e a base de teste contém {} elemento.'.format(
    treino_x.shape[0], teste_x.shape[0]))
    print(100* '-')
# ajustando modelos com a base de teste
    modelo = classificador
    modelo.fit(treino_x,treino_y)
    
    
# ------------------------------------ Validação -------------------------------------------------------

# matriz de confusão
    plt.style.use('default')
    matriz_confusao = plot_confusion_matrix(modelo, teste_x, teste_y, cmap='Blues', values_format='0.3g')
    plt.title('Matriz de Confusão')
# classification report
    previsoes = modelo.predict(teste_x)
    print(100* '-')
    print(classification_report(teste_y, previsoes))
    
# curva ROC e AUC
    prob_previsao = modelo.predict_proba(teste_x)[:, 1]
    tfp, tvp, limite = roc_curve(teste_y, prob_previsao)
    print('                        roc_auc', round(roc_auc_score(teste_y, prob_previsao),1))
    plt.style.use('seaborn')
    # Configurações do layout
    fig, ax = plt.subplots(figsize=(5, 5))
    sns.lineplot(tfp, tvp)
    # setando titulo do gráfico
    ax.set_title("Cruva ROC.", size=14)
    # setando titulo eixo x
    ax.set_xlabel("Especifidade", fontsize=14)
    # setando titulo eixo y
    ax.set_ylabel("Sensibilidade", fontsize=14)
    plt.plot([0, 1], ls="--", c = 'red') #plotando linha guia pontilhada vermelha
    plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') #plotando linha guia pontilhada verde
    plt.show()

    return modelo, matriz_confusao
```

```python
modelo(dados, LogisticRegression(max_iter=1000))
```
