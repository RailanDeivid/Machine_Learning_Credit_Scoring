## Separando VariÃ¡veis

 A base de dados que estamos trabalhando atÃ© agora Ã© inspirada em um conjunto de dados alemÃ£o e, como podemos perceber, temos uma Ãºnica variÃ¡vel que reÃºne informaÃ§Ãµes de sexo e de estado civil.  Acontece aqui, no Brasil, os modelos de crÃ©ditos sÃ£o fiscalizados e regulamentados pelo Banco Central e existem diversas regras que precisam ser seguidas e existem informaÃ§Ãµes que nÃ£o podem ser associadas ao modelo.
 
 
 Essas informaÃ§Ãµes sÃ£o as chamadas informaÃ§Ãµes sensÃ­veis. O que nÃ³s consideramos aqui, no Brasil, o que o Banco Central considera uma informaÃ§Ã£o sensÃ­vel? Origem social e Ã©tnica, condiÃ§Ãµes de saÃºde, informaÃ§Ã£o genÃ©tica, orientaÃ§Ã£o sexual, convicÃ§Ãµes polÃ­ticas, religiosas e filosÃ³ficas.

EntÃ£o o sexo Ã© uma informaÃ§Ã£o sensÃ­vel e o estado civil nÃ£o Ã©. Pelo Banco Central, nÃ³s nÃ£o podemos fazer a diferenciaÃ§Ã£o na hora da concessÃ£o de crÃ©dito para diferentes sexos. Cada empresa pode decidir polÃ­ticas de crÃ©dito diferentes, mas a concessÃ£o, o modelo de concessÃ£o, nÃ£o pode ter como argumento o sexo. Voltando para a nossa base de dados, nÃ³s temos usado atÃ© agora o sexo e estado civil aqui.


**E diante disso, iremos tratar essas variÃ¡veis.**

* Categorias da variÃ¡vel 'sexo_est_civil'
        male: divorced/separated = 1
        female: divorced/separated/married = 2
        male: single = 3
        male: married/widowed = 4
        female: single =5
        
* Nova variÃ¡vel 'sexo'
        1 = mulher

## Descartando colunas

```python
colunas_removidas = ['sexo_est_civil', 'sexo', 'idade', 'prazo_emprestimo_meses', 'valor_emprestimo']
dados = df.drop(columns=colunas_removidas)
dados.head()
```

# Criando bases de treino e teste

```python
y = dados['default']
x = dados.drop(columns='default')
```

```python
from sklearn.model_selection import train_test_split

SEED = 77
treino_x, teste_x, treino_y, teste_y = train_test_split(
    x, y, test_size=0.30, random_state=SEED)
```

```python
print('A base de treino contÃ©m {} elementos e a base de teste contÃ©m {} elemento.'.format(
    treino_x.shape[0], teste_x.shape[0]))
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter=1000)
modelo.fit(treino_x, treino_y)

# acuracia de treino
print(modelo.score(treino_x, treino_y))
```

# Verificando as MÃ©tricas

 cross-validation Ã© uma tÃ©cnica para para saber quÃ£o bem o seu modelo estÃ¡ sendo generalizado pelo modelo supervisionado (a regressÃ£o logÃ­stica Ã© um desses modelos) produzido. VocÃª pode se perguntar se a separaÃ§Ã£o que fizemos da base em teste jÃ¡ nÃ£o cumpre esta funÃ§Ã£o. E a resposta Ã© que sim, a base de teste jÃ¡ serve como avaliaÃ§Ã£o para o seu modelo.

Entretanto, o cross-validation cria diferentes amostras da base de dados para treino e teste e assim, vocÃª pode ter certeza da sua performance. Existem diversas tÃ©cnicas para fazer isso, mas aqui, vamos focar apenas em uma tÃ©cnica: o K-Fold.

Esta tÃ©cnica consiste em dividir sua base K vezes, realizar o treino com a maior parte destas amostras e depois testar a Ãºltima parte, geralmente.

## AcurÃ¡cia de teste

```python
previsoes = modelo.predict(teste_x)
previsoes
```

ele me devolve um array de 0 e 1. O que significa isso?

Significa que ele pegou a nossa base de teste e, a partir do modelo, ele me disse se aquela base terÃ¡ default 0 ou default 1. Essas previsÃµes serÃ£o usadas como parÃ¢metro para calcularmos, finalmente, a nossa acurÃ¡cia. Eu vou usar, para calcular a acurÃ¡cia, o accuracy score, que Ã© uma funÃ§Ã£o prÃ©-programada do Scikit-learn.

```python
from sklearn.metrics import accuracy_score

acuracia = round(accuracy_score(teste_y, previsoes), 2) *100
acuracia
```

Ela exibe a distribuiÃ§Ã£o dos registros em termos de suas classes atuais e de suas classes previstas. Isso indica a qualidade do modelo atual. Um modelo pode conter duas ou mais classes previstas.

## Classification report

```python
from sklearn.metrics import classification_report

print(classification_report(teste_y, previsoes))
```

* **Precision:**

     ProporÃ§Ã£o de prediÃ§Ãµes corretas de uma categoria em relaÃ§Ã£o a todas as prediÃ§Ãµes desta categoria.
         PrecisÃ£o = VP / VP + FP
     
     Ou seja, de todos os dados avaliados quantos relamente o modelo acertou.
     
         
* **Recall:**
    
    ProporÃ§Ã£o de prediÃ§Ãµes corretas de categoria alvo.
        Recall = VP / VP + FN
    
* **F1-Score:**

    MÃ©dia harmÃ´nica entre precisÃ£o e recall.
        F1 - score = 2 * (PrecisÃ£o * Recall) / (PrecisÃ£o + Recall)


* **Support:**

    Quantidade absoluta de casos em cada categoria.
        

## Plotando a curva ROC

```python
from sklearn.metrics import roc_curve, roc_auc_score

prob_previsao = modelo.predict_proba(teste_x)[:, 1]

prob_previsao
```

ele nos trÃ¡s uma array com a probabilidade do meu evento ser predito como 1 ou como 0. Isso serÃ¡ importante e serÃ¡ necessÃ¡rio como argumento para calcular a nossa curva ROC.

* Setando os parametros para a plotagem da curva ROC.

```python
#tfp = taxa de falsos positivos
#tvp = taxa de verdadeiros positivos

tfp, tvp, limite = roc_curve(teste_y, prob_previsao)


""" 
    A roc curve vai desenhar, vai definir qual serÃ¡ a minha curva ROC.
    A roc auc score calcula qual Ã© a Ã¡rea sob a curva ROC. 
    Ele calcula, atravÃ©s de uma integral, qual serÃ¡ a minha Ã¡rea sob a minha curva ROC.

"""
 
print('                   roc_auc', round(roc_auc_score(teste_y, prob_previsao),1))

import seaborn as sns
plt.style.use('seaborn')
# ConfiguraÃ§Ãµes do layout
fig, ax = plt.subplots(figsize=(5, 5))

sns.lineplot(tfp, tvp)
# setando titulo do grÃ¡fico
ax.set_title("Cruva ROC.", size=14)
# setando titulo eixo x
ax.set_xlabel("Especifidade", fontsize=14)
# setando titulo eixo y
ax.set_ylabel("Sensibilidade", fontsize=14)

plt.plot([0, 1], ls="--", c = 'red') #plotando linha guia pontilhada vermelha
plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') #plotando linha guia pontilhada verde


fig.tight_layout()
```

* area > 0.8, entao discriminacao excelente!

A curva ROC Avalia o desempenho do modelo a partir da relaÃ§Ã£o entre a taxa de verdadeiros positivos (sensibilidade) e a taxa de falsos positivos (especificidade) para diferentes limites.


Especificidade = VN / (VN + FP)  
Sensibilidade = VP (VP + FN)  

  O que significam essas duas mÃ©tricas? 
          
   A sensibilidade indica a taxa de casos em que a categoria alvo, ou seja, a categoria 1, foi classificada corretamente.  
   JÃ¡ a especificidade mede a taxa em que a categoria 0 foi, de maneira errada, classificada como 1


* **AUC:**

     Quantifica a Curva ROC atravÃ©s da Ã¡rea sob esta curva

    ConvenÃ§Ã£o EstatÃ­stica:

    - Ã¡rea < 0,7: baixa discriminiaÃ§Ã£o
    - 0,7 â‰¤ Ã¡rea < 0,8: discriminaÃ§Ã£o aceitÃ¡vel
    - 0,8 â‰¤ Ã¡rea < 0,9: discriminaÃ§Ã£o excelente
    - Ã¡rea > 0,9: discriminaÃ§Ã£o excepcional

* remoÃ§Ã£o de colunas da base de dados

# Ajuste do modelo

* Criando um funÃ§Ã£o para modelagem

```python
def modelo(dados, classificador):  
# separando base de modelagem e variÃ¡vel resposta
    y = dados['default']
    x = dados.drop(columns='default')
# separando dados de treino e teste
    SEED = 77
    treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size=0.30, random_state=SEED)
    base_treino = treino_x.shape[0]
    base_teste =  teste_x.shape[0]
    print('A base de treino contÃ©m {} elementos e a base de teste contÃ©m {} elemento.'.format(
    treino_x.shape[0], teste_x.shape[0]))
    print(100* '-')
# ajustando modelos com a base de teste
    modelo = classificador
    modelo.fit(treino_x,treino_y)
    
    
# ------------------------------------ ValidaÃ§Ã£o -------------------------------------------------------

# matriz de confusÃ£o
    plt.style.use('default')
    matriz_confusao = plot_confusion_matrix(modelo, teste_x, teste_y, cmap='Blues', values_format='0.3g')
    plt.title('Matriz de ConfusÃ£o')
# classification report
    previsoes = modelo.predict(teste_x)
    print(100* '-')
    print(classification_report(teste_y, previsoes))
    
# curva ROC e AUC
    prob_previsao = modelo.predict_proba(teste_x)[:, 1]
    tfp, tvp, limite = roc_curve(teste_y, prob_previsao)
    print('                        roc_auc', round(roc_auc_score(teste_y, prob_previsao),1))
    plt.style.use('seaborn')
    # ConfiguraÃ§Ãµes do layout
    fig, ax = plt.subplots(figsize=(5, 5))
    sns.lineplot(tfp, tvp)
    # setando titulo do grÃ¡fico
    ax.set_title("Cruva ROC.", size=14)
    # setando titulo eixo x
    ax.set_xlabel("Especifidade", fontsize=14)
    # setando titulo eixo y
    ax.set_ylabel("Sensibilidade", fontsize=14)
    plt.plot([0, 1], ls="--", c = 'red') #plotando linha guia pontilhada vermelha
    plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') #plotando linha guia pontilhada verde
    plt.show()

    return modelo, matriz_confusao
```

```python
modelo(dados, LogisticRegression(max_iter=1000))
```
